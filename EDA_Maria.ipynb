```python
# названия предварительно непрерывных значений колонок
name_continous = df.columns[df.max() - df.min() > 1]
```

```python
# количество нулей в каждой колонке
counts_zero_to_columns = (df[name_continous] == 0).sum(axis=0)
print(counts_zero_to_columns.sort_values())
```

```python
# гистограмма распределение количества нулей 
plt.hist(df[name_continous].shape[0] - df[name_continous][df[name_continous] != 0].count(), bins = 12, color='#A115FE', edgecolor = 'black')

plt.title('Распределение столбцов по количеству значений, равных нулю')
plt.xlabel('Количество зачений, которые равняются нулю в колонках')
plt.ylabel('Количество столбцов')

plt.show()
```
